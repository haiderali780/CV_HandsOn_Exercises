```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def negative_transformation(img):
    original_img = cv.imread(img, cv.IMREAD_GRAYSCALE)
    L = 2**(original_img.itemsize * 8)
    negative_img = L - 1 - original_img

    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(original_img, cmap='gray')
    axes[0].set_title("Original Image")
    
    axes[1].imshow(negative_img, cmap='gray')
    axes[1].set_title("Negative Transformed Image")
    
    plt.show()

negative_transformation("img.jpg")
```

```python
def logTransformation(img):
    base = int(input("Enter Base of logarithm"))
    image = cv.imread(img, cv.IMREAD_GRAYSCALE)
    max_val = np.max(image)
    log_image = np.log1p(image)
    log_image = (log_image / np.max(log_image)) * 255
    log_image = np.round(log_image).astype(np.uint8)
    
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(image, cmap='gray')
    axes[0].set_title("Original Image")
    axes[1].imshow(log_image, cmap='gray')
    axes[1].set_title("Log Transformed Image")
    plt.show()

logTransformation("img.jpg")
```

```python
def power_transformation(img, c=1):
    gamma = float(input("Enter gamma value: "))
    original_img = cv.imread(img, cv.IMREAD_GRAYSCALE)
    normalized_img = original_img / 255.0
    gamma_corrected_img = c * (255 * (normalized_img ** gamma))
    gamma_corrected_img = np.round(gamma_corrected_img).astype(np.uint8)

    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(original_img, cmap='gray')
    axes[0].set_title("Original Image")
    axes[1].imshow(gamma_corrected_img, cmap='gray')
    axes[1].set_title("Gamma Corrected Image")
    plt.show()

power_transformation("img.jpg")
```

```python
def contrast_stretching(img):
    min_gray = int(input("Minimum Gray Scale Level: "))
    max_gray = int(input("Maximum Gray Scale Level: "))

    original_img = cv.imread(img, cv.IMREAD_GRAYSCALE)
    min_original = np.min(original_img)
    max_original = np.max(original_img)

    contrast_stretched_img = ((original_img - min_original) / (max_original - min_original)) * (max_gray - min_gray) + min_gray
    contrast_stretched_img = np.clip(contrast_stretched_img, min_gray, max_gray).astype(np.uint8)

    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(original_img, cmap='gray')
    axes[0].set_title("Original Image")
    axes[1].imshow(contrast_stretched_img, cmap='gray')
    axes[1].set_title("Contrast Stretched Image")
    plt.show()

contrast_stretching("img.jpg")
```

```python
def gray_level_slicing(img):
    min_gray = int(input("Enter minimum gray scale: "))
    max_gray = int(input("Enter maximum gray scale: "))

    original_img = cv.imread(img, cv.IMREAD_GRAYSCALE)
    gray_level_sliced_img = np.where((original_img >= min_gray) & (original_img <= max_gray), 255, original_img)

    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(original_img, cmap='gray')
    axes[0].set_title("Original Image")
    axes[1].imshow(gray_level_sliced_img, cmap='gray')
    axes[1].set_title("Gray Level Sliced Image")
    plt.show()

gray_level_slicing("img.jpg")
```

```python
def bitSlicing(img):
    originalImg = cv.imread(img, cv.IMREAD_GRAYSCALE)
    bits = int(input("How many bits to be extraced"))

    if(bits>=0 and bits<=7):
        bitPlaneImg = (originalImg>> bits) & 1
        bitPlaneImg =255* bitPlaneImg

        fig,axes = plt.subplots(1,2,figsize=(10,5))

        axes[0].imshow(originalImg,cmap='gray')
        axes[0].set_title("Original Image")
        axes[1].imshow(bitPlaneImg,cmap='gray')
        axes[1].set_title("Bit Plane Sliced Image")
        plt.show()
    else:
        return 
bitSlicing("img.jpg")
    

```
